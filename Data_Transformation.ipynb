```python
# Q1: Handle Missing Values

import pandas as pd

df = pd.read_csv("/content/employee_productivity_dataset.csv")

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median(), inplace=True)
df['Performance_Score'].fillna(df['Performance_Score'].mean(), inplace=True)

print("Missing Values Handled")
print(df.head())


```

```python
# Q2: Label Encoding (Gender, Department)

from sklearn.preprocessing import LabelEncoder

for col in ['Gender', 'Department']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

print("Label Encoding Done")
print(df[['Gender','Department']].head())
```

```python
# Q3: One-Hot Encoding (Work_Mode, Location)
df = pd.get_dummies(df, columns=['Work_Mode','Location'], drop_first=True)

print("One-Hot Encoding Done")
print(df.head())
print("New columns created:", df.shape[1])
```

```python
# Q4: Normalization (MinMaxScaler on Salary, Hours_Worked_Per_Week)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['Salary','Hours_Worked_Per_Week']] = scaler.fit_transform(df[['Salary','Hours_Worked_Per_Week']])

print("Normalization Done")
print(df[['Salary','Hours_Worked_Per_Week']].head())
```

```python
# Q5: Standardization (StandardScaler on Age)
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
df[['Age']] = std_scaler.fit_transform(df[['Age']])

print("Standardization Done")
print(df[['Age']].head())
```

```python
# Q6: Compare Scaling Methods (Salary)
salary_data = df[['Salary']]

# MinMax
minmax = MinMaxScaler()
salary_minmax = minmax.fit_transform(salary_data)

# StandardScaler
std = StandardScaler()
salary_std = std.fit_transform(salary_data)

comparison = pd.DataFrame({
    'Original': salary_data['Salary'],
    'MinMax': salary_minmax.flatten(),
    'Standard': salary_std.flatten()
})

print("Compare Scaling Methods")
print(comparison.head())
```

```python
# Q7: Build Preprocessing Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numeric_features = ['Age','Salary','Hours_Worked_Per_Week','Performance_Score','Projects_Completed']
categorical_features = ['Gender','Department']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Q8: Apply Pipeline
df_processed = preprocessor.fit_transform(df)

print("Pipeline Applied")
print("Transformed dataset shape:", df_processed.shape)
```
